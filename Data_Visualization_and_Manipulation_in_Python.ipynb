#**Part-1: Data Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
tips_data = pd.read_csv("tips.csv")

# printing the top 10 rows
display(tips_data.head(10))
```

##**Scatter Plot**

```python
plt.scatter(tips_data['total_bill'], tips_data['tip'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# plt.colorbar()

plt.show()
```

##**Linear Regression**



```python
from sklearn.linear_model import LinearRegression

regressor=LinearRegression()
type(tips_data['total_bill'])
X = tips_data.iloc[:, :1]  # independent
y = tips_data.iloc[:, 1:2]  # dependent

## Here the first parameter in iloc[] will take the rows' range, and the second one is the column range

regressor.fit(X, y)
# print(X, type(X))
```

In the Python Pandas library, .iloc[] is an indexer used for integer-location-based indexing of data in a DataFrame. It allows users to select specific rows and columns by providing integer indices, making it a valuable tool for data manipulation and extraction based on numerical positions within the DataFrame. This indexer is particularly useful when you want to access or manipulate data using integer-based positional indexing rather than labels.

```python
y_pred = regressor.predict(X)     # predicted value of y_test
plt.plot(X, y_pred, color='green')
plt.scatter(X, y)

plt.title('Total Bill vs Tip Regression Line')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(['Predicted Data', 'Actual Data'], title = 'Bill/Tip', loc='best', facecolor='white')

plt.show()
```

##**Polynomial Regression**

```python
salary_data = pd.read_csv("/content/Position_Salaries.csv")

print(salary_data.head(10))
```

```python
X = salary_data.iloc[:, 1:2]  # independent
y = salary_data.iloc[:, 2:3]  # dependent

plt.scatter(X, y)
plt.title('Level vs Salary')
plt.xlabel('Level')

plt.ylabel('Salary')
```

```python
from sklearn.preprocessing import PolynomialFeatures


X = salary_data.iloc[:, 1:2]  # independent
y = salary_data.iloc[:, 2:3]  # dependent


poly = PolynomialFeatures(degree=4)

X_poly = poly.fit_transform(X)
lr_2 = LinearRegression()
lr_2.fit(X_poly, y)

y_pred_poly = lr_2.predict(X_poly)
plt.plot(X, y_pred_poly, color='green')
plt.scatter(X, y)


plt.title('Level vs Salary')
plt.xlabel('Level')

plt.ylabel('Salary')
plt.legend(['Predicted Data', 'Actual Data'], title = 'level/salary', loc='best', facecolor='white')

plt.show()
```

#**Part-2: Data Manipulation**

```python
flights_data = pd.read_csv("/content/flights.csv")

len(flights_data)
flights_data.head(10)

# flights_data.iloc[335770:335776]
```

```python
flights_data.describe()
```

```python
flights_data.groupby("month").count()
```

```python
distance_interval = 1000
array_length = (max(flights_data['distance']) - min(flights_data['distance'])) // distance_interval + 1
print(array_length, min(flights_data['distance']), max(flights_data['distance']))
```

```python
flight_count_on_distance = [0] * array_length

for _, row in flights_data.iterrows():
  list_index = row['distance'] // distance_interval
  flight_count_on_distance[list_index] += 1

print(flight_count_on_distance)
```

```python
lower_value = 1
upper_value = 1000

distance_label = []
for i in range(array_length):
  distance_label.append(str(lower_value) + "-" + str(upper_value))
  lower_value += 1000
  upper_value += 1000

print(distance_label)
```

```python
plt.pie(flight_count_on_distance, labels=distance_label, autopct='%1.1f%%')
plt.title('Flights per Distance')
plt.legend()
plt.show()
```
